```python
import ast
```

```python
def convert(str):
    L = []
    for i in ast.literal_eval(str):
        L.append(i['name']) 
    return L 
```

```python
data['genres'] = data['genres'].apply(convert)
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
data['cast'] = data['cast'].apply(convert)
data.head()
```

```python
data['cast'] = data['cast'].apply(lambda x:x[0:3])
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
data['crew'] = data['crew'].apply(fetch_director)
```

```python
data.sample(5)
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
data['cast'] = data['cast'].apply(collapse)
data['crew'] = data['crew'].apply(collapse)
data['genres'] = data['genres'].apply(collapse)
```

```python
data.head()
```

```python
data['overview'] = data['overview'].apply(lambda x:x.split())
```

```python
data['tags'] = data['overview'] + data['genres'] + data['cast'] + data['crew']
```

```python
new = data.drop(columns=['overview','genres','cast','crew'])
```

```python
new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
new['tags'] = new['tags'].apply(
    lambda x: ", ".join(
        i['name'].lower() if isinstance(i, dict) and 'name' in i else str(i).lower()
        for i in x
    ) if isinstance(x, list) else str(x).lower()
)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import TruncatedSVD
import pandas as pd

#  Vectorize text data
cv = CountVectorizer(max_features=22283, stop_words='english')
vectors = cv.fit_transform(new['tags'])  # 'new["tags"]' contains text

print("Original shape:", vectors.shape)  # (44566, 30000)


svd = TruncatedSVD(n_components=2283, random_state=69)
vector = svd.fit_transform(vectors)

print("Reduced shape:", vector.shape)  # (44566, 3000)
```

```python
print(svd.explained_variance_ratio_.sum())
```

```python
print(f"Explained variance: {svd.explained_variance_ratio_.sum():.2f}")
```

```python
import matplotlib.pyplot as plt

svd_2d = TruncatedSVD(n_components=2)
vectors_2d = svd_2d.fit_transform(vectors)

plt.scatter(vectors_2d[:, 0], vectors_2d[:, 1], s=1)
plt.title("2D Projection of Tags")
plt.xlabel("SVD1")
plt.ylabel("SVD2")
plt.show()
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
recommend('Jumanji')
```
